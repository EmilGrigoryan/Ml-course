```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('weatherAUS.csv')
```

```python
# подготовка данных
```

```python
data
```

```python
data.isna().sum()
```

```python
data = data.drop(['Date', 'Evaporation', 'Sunshine', 'WindGustDir', 'WindDir9am', 'WindDir3pm'], axis=1)
```

```python
data = data[data['RainTomorrow'].notna()]
```

```python
for i, v in enumerate(data['Location'].unique()):
    data = data.replace(v, i)
    
```

```python
data = data.replace('No', 0)
data = data.replace('Yes', 1)
```

```python
data
```

```python
data.columns
```

```python
data = data.fillna(data.mean())
```

```python
data
```

```python
data.isna().sum()
```

```python
y = data['RainTomorrow']
```

```python
X = data.drop('RainTomorrow', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)
```

```python
# Байес
```

```python
gnb = GaussianNB()
```

```python
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
# Логистическая регрессия 
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
y_pred_log = clf.predict(X_test)
```

```python
print(y_pred_log)
```

```python
accuracy_score(y_pred_log, y_test)
```

```python
# KNN
```

```python
clf2 = KNeighborsClassifier(n_neighbors=3)
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred_k3 = clf2.predict(X_test)
```

```python
y_pred_k3
```

```python
accuracy_score(list(y_pred_k3), y_test)
```

```python
clf3 = KNeighborsClassifier(n_neighbors=5)
```

```python
clf3.fit(X_train, y_train)
```

```python
y_pred_k5 = clf3.predict(X_test)
```

```python
accuracy_score(list(y_pred_k5), y_test)
```

Лучший результат показала логистическая регрессия
